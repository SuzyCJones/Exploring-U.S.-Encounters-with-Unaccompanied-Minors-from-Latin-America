## Wrangle Columns: 

```r
library("tidyr")
```

### Combine Month and Year columns to make a date column

```r
BPEncounters1 <- unite(BPEncounters, Date, Month..abbv., Fiscal.Year, sep = " ")
```

```r
head(BPEncounters1)
```

### Renaming Columns: 

```r
names(BPEncounters1)[names(BPEncounters1) == "Citizenship.Grouping"] <- "Citizenship"
```

```r
names(BPEncounters1)[names(BPEncounters1) == "Title.of.Authority"] <- "Law.Enforced"
```

```r
head(BPEncounters1)
```

### Remove "month grouping"

```r
BPEncounters1 <- BPEncounters1[, -which(names(BPEncounters1) == "Month.Grouping")]
```

```r
head(BPEncounters1)
```

## Time to graph!!!!

### drawing conclusions between the timeframes and encounter counts

```r
library ("ggplot2")
```

```r
BPEncountersCountPlot <- ggplot(data = BPEncounters1, aes(x = Encounter.Count, y = Date))
```

```r
BPEncountersCountPlot <- BPEncountersCountPlot + geom_bar(stat = "sum", fill = "steelblue")
```

```r
print(BPEncountersCountPlot)
```
