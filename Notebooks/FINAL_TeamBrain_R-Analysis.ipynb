# Page 1 - Abstract <a class="anchor" id="PAN_page_1"></a>

<hr style="height:5px;border-width:0;color:MediumAquamarine;background-color:#008080">

### Background
Even though the cause of autism spectrum disorders (ASD) remains unknown, the current understanding points towards complex interactions between environmental and genetic factors. One important environmental factor to consider is intake of toxic and essential elements, and their role in metabolism. Essential elements have received considerably less attention in the literature than the presence of toxins in urine.

- [Mendeley Reference](https://data.mendeley.com/datasets/79fk29dvk6/1)
- [Science Direct Reference](https://www.sciencedirect.com/science/article/abs/pii/S1750946720300088?via%3Dihub)

### Method
The purpose of this investigation is to comprehensively assess the association between urinary element compositions of 28 mothers who had young children with ASD and 29 mothers who had young typically developing (TD) children, and in a subset of their children (21 with ASD and 26 TD).

### Hypothesis 
We expect to find an increase in toxic elements and a decrease in essential elements for children and mothers with Autism Spectrum Disorder status compared to typically developing children and mothers.   

### Objectives
1. Is there a predictive relationship between levels of toxic and essential elements and the occurrence of ASD in children?
2. From the modeled elements, how does the urine test differ (Toxic vs Essential) for kids with Autism (ASD) to Typically Developing kids (TD)?
3. Is there a predictive relationship between levels of toxic and essential elements and the occurrence of ASD in mothers?
3. From the modeled elements, how does the urine test differ (Toxic vs Essential) for mothers with Autism (ASD) to Typically Developing moms (TD)?
4. From the modeled elements, how does the urine test differ (Toxic vs Essential) for kids with Autism (ASD) to their moms with Autism(ASD)?

### Analysis Setup
DV1 = Status with 2 levels (ASD or TD)<br>
DV2 = Type with 2 levels (Child or Mother)<br>
IV1 = Group of 20 Toxic elements (Aluminum to Uranium) <br>
IV2 = Group of 19 Essential elements (Boron to Creatinine)

# Logistics Regression to test data accuracy of predicting ASD or TD for childToxic

```r
# Fit logistic regression model with all 20 predictor variables
mylogitToxic <- glm(Status ~ ., data = childToxic2, family = binomial())
summary(mylogitToxic)
```

```r
childToxic3 <- childToxic2 %>% 
dplyr::select_if(is.numeric)

probabilities <- predict(mylogitToxic, type = "response")

probabilities <- predict(mylogitToxic, type = "response")
childToxic3$Predicted <- ifelse(probabilities > .5, "pos", "neg")

#Recode the new Predicted Variable
childToxic3$PredictedR <- NA
childToxic3$PredictedR[childToxic3$Predicted=='pos'] <- 1
childToxic3$PredictedR[childToxic3$Predicted=='neg'] <- 0

childToxic3$Status <- as.factor(childToxic2$Status)
childToxic3$PredictedR <- as.factor(childToxic3$PredictedR)
```

```r
#Distribution of Status Variable
table(childToxic2$Status)
```

```r
### Create a Confusion Matrix to test the accuracy of the prediction
conf_mat <- caret::confusionMatrix(childToxic3$PredictedR, childToxic3$Status)
conf_mat
```

```r
levels(childToxic3$PredictedR)  # Check levels of PredictedR
levels(childToxic3$Status)  # Check levels of Status
```

```r
#Calucate and assess the linearity of the logit
childToxic3 <- childToxic2 %>% 
dplyr::select_if(is.numeric)

predictors <- colnames(childToxic3)
```

```r
#create the logit
childToxic3 <- childToxic3 %>%
mutate(logit=log(probabilities/(1-probabilities))) %>%
gather(key= "predictors", value="predictor.value", -logit)
```

```r
infl <- influence.measures(mylogitToxic)
summary(infl)
```

```r
# Fit a glm model
model <- glm(Status ~ ., data = childToxic2, family = binomial())

#Review logistics regression summary
summary(mylogitToxic)

# Extract just the coefficients
coefficients <- coef(summary(model))
print(coefficients)
```

Since the coefficients are not significantly different from 0, Aluminum, Antimony, Arsenic, Barium, Bismuth, Cadmium, Cesium, Lead, Nickel, Thallium, Tin and Uranium are not significant predictors of ASD. <br>
Tungsten is the only significant predictor with a significant p-value and an estimate of 2.13
